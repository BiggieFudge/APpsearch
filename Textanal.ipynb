```python
#Imports
import pandas as pd
import time
import hebrew_tokenizer as ht
```

```python
def tokenizer(str):
    
    print(type(str))
    if type(str) is "str":
        print("!")
        return ""
    list = []
    tokens = ht.tokenize(str)
    for grp, token, token_num, (start_index, end_index) in tokens:
        print(token, grp)
        if grp == ("HEBREW" or "NUMBER"):
            list.append(token)
    print(list)
    return list
```

```python
og_df = pd.read_csv("facebook_scrape.csv")
og_df.insert(len(og_df.columns),column="tokenized", value="")
print(og_df.columns)
for row in og_df.index:
    print(og_df["text"][row])
    og_df.at[row,"tokenized"] = tokenizer(og_df["text"][row])
```
